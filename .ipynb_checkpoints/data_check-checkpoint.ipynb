Have to replace multiple poorly named tags with actual named tags. <br>
Chosen to replace: <br>
E, EM, EN, , EN:, ENEOS, HEN#, en, NE, RN, MIX, ENC with 'EN'. <br>
HII, HIS, USN with HI. <br>
U, UI, UN., UNEOS, _, uN, un with UN. <br>

```python
for_EN = ['E', 'EM', 'EN,' , 'EN:', 'ENEOS', 'HEN#', 'en', 'NE', 'RN', 'MIX', 'ENC']
for pos, tag in enumerate(tags):
    if tag in for_EN: tags[pos] = 'EN'
```

```python
for_HI = ['HII', 'HIS', 'USN']
for pos, tag in enumerate(tags):
    if tag in for_HI: tags[pos] = 'HI'
```

```python
for_UN = ['U', 'UI', 'UN.', 'UNEOS', '_', 'uN', 'un']
for pos, tag in enumerate(tags):
    if tag in for_UN: tags[pos] = 'UN'
```

```python
[s for i, s in enumerate(words) if tags[i] == 'MI']
```

```python
counter = 0
for i, s in enumerate(words):
    if tags[i] == 'MI':
        if counter == 0:
            tags[i] = 'HI'
            counter += 1
        else:
            tags[i] = 'EN'
```
